## Задание № 1

График № 4. Точечная диаграмма для признаков depth и price показывает отношение глубины алмаза к его цене. Разной формы показываем степень огранки алмаза


```python
#Точечная диаграмма 
sns.scatterplot(data=df, x="price", y="depth", color='black', style='cut')
plt.title('Точечная диаграмма по признакам depth и price')
plt.show()
```

График № 8

Реализовано через Seaborn, хотя более похож график из ggplot2 с его типами графиков: geom_point + geom_smooth. Для реализации надо подключать новый блокнот с языком R. <br>
На точечной диаграмме с линиями регрессии, показаны зависимость цены алмазов от их карата

```python
# Создание графика
plt.figure(figsize=(10, 6))
# sns.scatterplot(data=df, x='carat', y='price', hue='cut', alpha=0.6)  # Точечная диаграмма
sns.lmplot(data=df, x='carat', y='price', hue='cut')  # Линия регрессии

# Настройка заголовков и подписей
plt.title("Цена алмазов в зависимости от карата")
plt.xlabel("carat")
plt.ylabel("price")
plt.legend(title="cut")
plt.show()  # Отображение графика
```

График № 4

Box plot — график, использующийся в описательной статистике, компактно изображающий одномерное распределение вероятностей. С помощью box plot визуализируем зависимость количества цилиндров от объема двигателя.

```python
sns.boxplot(data=mt, x='cyl', y='disp', hue='am', palette='Set2');
plt.title('Box plot по признакам количество цилиндров и рабочий объем двигателя')
plt.show()
```

График № 8

Dot Plot 
Иногда для случаев, когда мы исследуем числовую переменную подходит простой график, который отображает распределение наших наблюдений на одной соответствующей числовой шкале.
На графике представлена зависимость количество цилиндров в двигателе от расходе топлива в милях на галлон

```python
sns.swarmplot(data=mt, x='mpg', hue='cyl')
plt.title('Данные о расходе топлива автомобилей в милях на галлон (miles per gallon)')
plt.show()
```
